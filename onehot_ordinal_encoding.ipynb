```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('final_dataset/final_model.csv')
```

```python
df.head()
```

```python
enc = OrdinalEncoder()
one = OneHotEncoder()
```

```python
# ordinal encoding for "Neighborhood" and "Premise"

df[["Neighborhood"]] = enc.fit_transform(df[["Neighborhood"]])
df[["Premise"]] = enc.fit_transform(df[["Premise"]])
```

```python
# onehot encoding for "Race" and "District"

df[["Race"]] = one.fit_transform(df[["Race"]])
df[["District"]] = one.fit_transform(df[["District"]])
```
