```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('final_dataset/multi_label_encoded.csv')
```

```python
# ordinal encoding for "Race" and "District"

enc = OrdinalEncoder()
df[["Race"]] = enc.fit_transform(df[["Race"]])
df[["District"]] = enc.fit_transform(df[["District"]])
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# splitting dataset into training and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# backward feature selection

backward_feature_selection = SequentialFeatureSelector(RandomForestClassifier(n_jobs = -1), 
                                                      k_features = (1, 19), 
                                                      forward = False,
                                                      floating = False,
                                                      verbose = 2,
                                                      scoring = "accuracy",
                                                      cv = 0).fit(X_train, y_train)
```
