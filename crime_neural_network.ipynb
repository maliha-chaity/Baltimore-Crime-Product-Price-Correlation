```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
import tensorflow as tf
```

```python
df = pd.read_csv('final_dataset/final_model6.csv')
```

```python
X = df.iloc[:, 0:16].values
y = df.iloc[:, -1].values
```

```python
# one hot encoding for "Race" and "District"

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [4, 5])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
# feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# artificial neural network

ann = tf.keras.models.Sequential()
```

```python
# adding input layer and the first hidden layer

ann.add(tf.keras.layers.Dense(units=32, activation='relu'))
```

```python
# adding second hidden layer

ann.add(tf.keras.layers.Dense(units=32, activation='relu'))
```

```python
# adding output layer

ann.add(tf.keras.layers.Dense(units=1, activation='softmax'))
```

```python
# compiling ANN

ann.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# training ANN on training set

ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```
