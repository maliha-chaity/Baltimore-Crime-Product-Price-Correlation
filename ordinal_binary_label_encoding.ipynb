```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('final_dataset/binary_classification.csv')
```

```python
# ordinal encoding on 'Neighborhood' and 'Premise'

enc = OrdinalEncoder()
df[["Neighborhood"]] = enc.fit_transform(df[["Neighborhood"]])
df[["Premise"]] = enc.fit_transform(df[["Premise"]])
```

```python
# label encoding on 'Description'

label_encoder = LabelEncoder()
df["Description"] = label_encoder.fit_transform(df["Description"])
```

```python
df.to_csv('binary_label_encoded.csv', index = True) 
```
